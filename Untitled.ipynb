```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_curve
import xgboost as xgb
```

```python
df = pd.read_csv("Fraud.csv")
df.head()
```

```python
df.info()
df.describe()
df["isFraud"].value_counts(normalize=True)  # check class imbalance
```

```python
df = df.drop(["nameOrig", "nameDest"], axis=1)
```

```python
pd.crosstab(df["type"], df["isFraud"], normalize='index')
```

```python
#Fraud does not happen in Payment
df = df[df["type"].isin(["TRANSFER", "CASH_OUT"])].copy()
```

```python
df["type"] = df["type"].map({"TRANSFER": 0, "CASH_OUT": 1})
```

```python
df["errorBalanceOrig"] = df["oldbalanceOrg"] - df["newbalanceOrig"] - df["amount"]
df["errorBalanceDest"] = df["newbalanceDest"] - df["oldbalanceDest"] - df["amount"]
```

```python
df.head()
```

```python
X = df.drop(["isFraud", "isFlaggedFraud"], axis=1)
y = df["isFraud"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
scale = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
print(f"Scale pos weight = {scale}")
```

```python
model = xgb.XGBClassifier(
    eval_metric='logloss',
    use_label_encoder=False,
    scale_pos_weight=336.34,
    max_depth=4,
    learning_rate=0.1,
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import precision_recall_curve

y_proba = model.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

plt.plot(thresholds, precision[:-1], label='Precision')
plt.plot(thresholds, recall[:-1], label='Recall')
plt.xlabel("Threshold")
plt.title("Precision vs Recall")
plt.legend()
plt.grid()
plt.show()
```

```python
xgb.plot_importance(model, max_num_features=10)
plt.title("Top Features")
plt.show()
```
